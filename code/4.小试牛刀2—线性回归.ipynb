# 利用随机梯度下降算法实现线性回归

## 1.导入相关的包和创建数据

```python
import torch as t
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
t.manual_seed(100)
```

```python
def get_fake_data(batch_size = 8):
    x = t.rand(batch_size, 1) * 20
    y = 2 * x + (1 + t.randn(batch_size ,1)) * 3  #y = x * 2 + 3
    
    return x, y
    
```

```python
x, y = get_fake_data()
print(x.shape)
print(y.shape)
```

**关于随机数的生成的一些区分**

1. t.rand是生成一个[0，1]之间均匀分布的随机数
2. t.randn是生成一个高斯分布的随机数

## 2.手写线性回归算法

**先推一下公式：**  

1）线性回归公式
$$\overline{y} = wX + b$$
2）损失函数公式
$$J=\frac{1}{2}*(\overline{y} - y )^2$$

3）求导公式：  
$$dw = \frac{\partial{J}}{\partial{y}}* \frac{\partial{y}}{\partial{w}}$$  

$$dJ = \overline{y} - y$$  

$$dy = X$$  

$$w_{i+1} = w_i - dw$$

```python
w = t.rand(1, 1)
b = t.zeros(1, 1)
print('init w:',w)
print('init b:',b)

lr = 0.001
for iter in range(2000):
    
    x, y = get_fake_data(200)
    
    #前向传播：计算y_predict预测值以及损失值
    y_predict = x.mm(w) + b.expand_as(y)
    
    loss = 1/2 * (y_predict - y ) ** 2
    loss = loss.sum()
    
    #反向传播:求导更新参数
    dJ = y_predict - y  #shape = (8,1)
    dy = x #shape = (8,1)
    dw = x.t().mm(dJ) # dJ * dy shape = (1,1)
    db = dJ.sum() * 1 # shape = (1,1)
    
    w.sub_( lr * dw)
    b.sub_( lr * db)

    print('w',w)
    print('b',b)
    
    if iter % 100==0:
        x_line = t.linspace(0,20.0,200).view(-1,1)#第一个维度自动补全
        y_line = x * w + b.expand_as(x)
        
        print(x_line.numpy().shape)
        print(y_line.numpy().shape)
        
        plt.plot(x_line.numpy(), y_line.numpy())
        plt.scatter(x.numpy(),y.numpy())
        plt.show()
        
print(w)
print(b)
    
```
