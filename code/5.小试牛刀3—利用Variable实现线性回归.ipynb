# 1.导入相关的包

```python
import torch as t
from torch.autograd import Variable as V
```

```python
%matplotlib inline
```

```python
from matplotlib import pyplot as plt
```

# 2.创建数据

```python
t.manual_seed(200)
```

```python
def get_fake_data(batch_size = 8):
    # get data : y = 2 * x + 3
    x = t.rand(batch_size , 1) * 20
    y = 2 * x + (t.randn(batch_size, 1) * 3) + 3
    return x, y
    
```

```python
x , y =get_fake_data(20)
plt.scatter(x.numpy(), y.numpy())
```

```python
x.shape
```

# 3.编写梯度下降算法，并绘制拟合曲线以及loss损失函数曲线

```python
w = V(t.rand(1, 1), requires_grad = True)
b = V(t.rand(1, 1),requires_grad = True)

lr = 0.01
x, y = get_fake_data(200)
x, y =V(x), V(y)
n = x.shape[0]
loss_history = []

iter_times = 200

for iter in range(iter_times):
    y_predict = x * w + b
    loss = 0
    loss = 0.5 * (y_predict - y) ** 2
    loss = loss.sum()/n
    loss_history.append(loss)
    
    loss.backward()
    
    w.data.sub_(lr * w.grad.data)
    b.data.sub_(lr * b.grad.data)
    
    w.grad.data.zero_()
    b.grad.data.zero_()

    if iter % (iter_times /10) ==0 :
        x_line = t.linspace(0,20.0,n).view(-1,1)
        y_line = x_line * w + b
        plt.plot(x_line.data.numpy(),y_line.data.numpy(),c = 'red')
        plt.scatter(x.numpy(), y.numpy())
        plt.show()
    
print(w,b)
loss_x_index = t.arange(0,iter_times)
plt.plot(loss_x_index.numpy(),loss_history,c = 'red')
plt.show()
```
