# 利用随机梯度下降算法实现线性回归

## 1.导入相关的包和创建数据

```python
import torch as t
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
t.manual_seed(100)
```

```python
def get_fake_data(batch_size = 8):
    x = t.rand(batch_size, 1) * 20
    y = 2 * x + (1 + t.randn(batch_size ,1)) * 3  #y = x * 2 + 3
    
    return x, y
    
```

```python
x, y = get_fake_data()
print(x.shape)
print(y.shape)
```

**关于随机数的生成的一些区分**

1. t.rand是生成一个[0，1]之间均匀分布的随机数
2. t.randn是生成一个高斯分布的随机数

## 2.手写线性回归算法

**先推一下公式：**  

1）线性回归公式
$$\overline{y} = wX + b$$
2）损失函数公式
$$J=\frac{1}{2}*(\overline{y} - y )^2$$

3）求导公式：  
$$dw = \frac{\partial{J}}{\partial{y}}* \frac{\partial{y}}{\partial{w}}$$  

$$dJ = \overline{y} - y$$  

$$dy = X$$  

$$w_{i+1} = w_i - dw$$

```python
w = t.rand(1, 1)
b = t.zeros(1, 1)
print('init w:',w)
print('init b:',b)

x, y = get_fake_data(200)
# x.shape = (200, 1)
# t.shape = (200, 1)
# w.shape = (1, 1)
# b.shape = (1, 1)


def forward(x, w, b):
    return  x * w + b #y_predict
    
def cost_function(y_predict, y):
    return 0.5 * (y_predict - y) ** 2


lr = 0.01
for iter in range(2000):
    
    #前向传播：计算y_predict预测值以及损失值
    y_predict = forward(x, w, b)
    
    n = x.shape[0]
    loss = cost_function(y_predict, y)
    loss = loss.sum() /n
    
    #反向传播:求导更新参数
    dJ = y_predict - y  #shape = (n,1)
    dw = (dJ.t() .mm(x)) / n  # dJ * dy shape = (1,1)
    db = dJ.sum() * 1 / n  # shape = (1,1)
    
    w.sub_( lr * dw)
    b.sub_( lr * db)


    
    if iter % 100==0:
        x_line = t.linspace(0,20.0,n).view(-1,1)#第一个维度自动补全
        y_line = x_line * w + b.expand_as(x)
        
#         print(x_line.numpy().shape)
#         print(y_line.numpy().shape)
        
        plt.plot(x_line.numpy(), y_line.numpy(),c = 'red')
        plt.scatter(x.numpy(),y.numpy())
        plt.show()
        
print(w)
print(b)
    
```
